```python
# Напишите код на Python, реализующий расчет длины вектора, заданного его координатами

import math
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
def vector_length(vector: list):
    if vector.__len__() < 2 or vector.__len__() > 3:
        raise Exception('Неверная размерность вектора')

    vector_length = 0

    for coordinate in vector:
        vector_length += coordinate * coordinate

    return math.sqrt(vector_length)
```

```python
a = [3, 2]
vector_length(a)
```

```python
### парабола
x = []
y1 = []
y2 = []

p = 5

for i in range(1000):
    x_ = i / 100
    x.append(x_)
    y1.append(math.sqrt(2 * p * x_))
    y2.append(-math.sqrt(2 * p * x_))
    
plt.plot(x, y1)
plt.plot(x, y2);
```

```python
# круг
plt.figure(figsize=(5, 5))

x, y, y1 = [], [], []
r, x0, y0 = 20, 3, 3

for x_ in np.linspace(-20, 20, 1000):
    x.append(x_ + x0)
    y.append(math.sqrt(r**2 - x_**2) + y0)
    y1.append(-math.sqrt(r**2 - x_**2) + y0)
    
plt.plot(x, y)
plt.plot(x, y1)
```

```python
# элипс
x, y, y1 = [], [], []
a, b = -20, 10

for x_ in np.linspace(-20, 20, 100):
    x.append(x_)
    y.append(math.sqrt(b**2 * (1 - (x_**2 / a**2))))
    y1.append(-math.sqrt(b**2 * (1 - (x_**2 / a**2))))

plt.plot(x, y)
plt.plot(x, y1)
```

```python
# гипербола

x, y, y1 = [], [], []
a, b = -2, 10

for x_ in np.linspace(-20, 20, 100):
    x.append(x_)
    y.append(math.sqrt(b**2 * (1 + (x_**2 / a**2))))
    y1.append(-math.sqrt(b**2 * (1 + (x_**2 / a**2))))

plt.plot(x, y)
plt.plot(x, y1)
```
